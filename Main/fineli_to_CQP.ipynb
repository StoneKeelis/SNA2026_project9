```python
import pandas as pd 
```

Creating a dataframe from Fineli dataset

```python
fineli_data = pd.read_csv("resultset.csv", sep=";")
```

```python
fineli_data.head()
```

Creating another dataframe

```python
names_for_korp = fineli_data.copy()
names_for_korp = names_for_korp[["name"]]
```

```python
names_for_korp.head()
```

Cleaning food names, removing spaces and duplicates

```python
names_for_korp = [x.lower().split(",")[0].split(" ")[0] for x in names_for_korp["name"]]

names_for_korp = list(dict.fromkeys(names_for_korp))
```

```python
print(names_for_korp)
```

Creating smaller sublists so that the Korp search does not crash

```python
names_for_korp1 = names_for_korp[0:150]
names_for_korp2 = names_for_korp[150:301]
names_for_korp3 = names_for_korp[301:451]
names_for_korp4 = names_for_korp[451:601]
names_for_korp5 = names_for_korp[601:751]
names_for_korp6 = names_for_korp[751:901]
names_for_korp7 = names_for_korp[901:1051]
names_for_korp8 = names_for_korp[1051:]
```

Modifying lists into Korp's CQP form

```python
CQP_search1 = " | ".join([f"\"{n}\"" for n in names_for_korp1])
CQP_search2 = " | ".join([f"\"{n}\"" for n in names_for_korp2])
CQP_search3 = " | ".join([f"\"{n}\"" for n in names_for_korp3])
CQP_search4 = " | ".join([f"\"{n}\"" for n in names_for_korp4])
CQP_search5 = " | ".join([f"\"{n}\"" for n in names_for_korp5])
CQP_search6 = " | ".join([f"\"{n}\"" for n in names_for_korp6])
CQP_search7 = " | ".join([f"\"{n}\"" for n in names_for_korp7])
CQP_search8 = " | ".join([f"\"{n}\"" for n in names_for_korp8])
```
